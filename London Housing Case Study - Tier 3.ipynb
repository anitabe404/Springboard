#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

Analyst's notes: There are three columns that are Unnamed and have no data (i.e. all the values in the column are null). These columns can be removed right away, as it is clear that they aren't boroughs.

```python
# Identify columns that start with Unnamed and contain only nulls
unnamed_cols = [col for col in properties if col.startswith('Unnamed: ') and properties[col].isna().sum() == len(properties)]

# Drop Unnamed columns with nulls from the dataframe
properties_named = properties.drop(columns=unnamed_cols)
properties_named.info()
```

_Analyst's notes: After dropping the Unnamed columns, there are 46 remaining columns in the dataframe, indicating that there are still columns listed that are not boroughs._

```python
# Define tuple of valid London Boroughs (including City of London)
LONDON_BOROUGHS = (
"City of London",
"Westminster",
"Kensington & Chelsea",
"Hammersmith & Fulham",
"Wandsworth",
"Lambeth",
"Southwark",
"Tower Hamlets",
"Hackney",
"Islington",
"Camden",
"Brent",
"Ealing",
"Hounslow",
"Richmond upon Thames",
"Kingston upon Thames",
"Merton",
"Sutton",
"Croydon",
"Bromley",
"Lewisham",
"Greenwich",
"Bexley",
"Havering",
"Barking & Dagenham",
"Redbridge",
"Newham",
"Waltham Forest",
"Haringey",
"Enfield",
"Barnet",
"Harrow",
"Hillingdon")

# Find columns that aren't in LONDON_BOROUGHS
is_not_borough = np.isin(properties_named.columns,LONDON_BOROUGHS,invert=True)
cols_not_borough = properties_named.columns[is_not_borough]
print(cols_not_borough)
```

_Analyst's notes: Unnamed: 0 is a valid column and should be kept in the dataframe. The other columns listed can be removed as they are not valid boroughs._

```python
# Create a new dataframe called properties_LB that only contains valid London Boroughs
# Note: Element 0 is a valid column even though it's not a borough, so it's retained.
properties_LB = properties_named.drop(columns=cols_not_borough[1:])
properties_LB.info()
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose dataframe and reset index.
properties_LB = properties_LB.T.reset_index()
properties_LB.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Fix first two values in first row in preparation for column renaming
properties_LB.loc[0,"index"] = "london_borough"
properties_LB.loc[0,0] = "postcode"

# Set column headings to value of first row
properties_LB = properties_LB.rename(columns=properties_LB.iloc[0])
properties_LB.head()
```

```python
# Now that the columns have been renamed, the first row can be dropped.
properties_LB = properties_LB.drop(0)
properties_LB.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# melted_prop = properties_2.melt(id_vars=['London Borough','Postcode'])
london_house_prices = pd.melt(properties_LB,id_vars=['london_borough','postcode'],var_name="year_month",value_name="avg_price")
# melted_prop = melted_prop.sort_values("London Borough")
london_house_prices.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
london_house_prices.info()
```

```python
# Convert avg_price column dtype to numeric
london_house_prices['avg_price'] = pd.to_numeric(london_house_prices['avg_price'])
london_house_prices.info()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(df,borough):
    borough_prices = df[df['london_borough'] == borough]
    avg_price_1998 = borough_prices[borough_prices['year'] == 1998]['avg_price'].mean()
    avg_price_2018 = borough_prices[borough_prices['year'] == 2018]['avg_price'].mean()
    return avg_price_2018 / avg_price_1998
```

_Analyst's notes: Using the suggested function, I am going to calculate the average price ratio for each borough and sort the list by price ratio._

```python
# Test create_price_ratio on Greenwich
print(create_price_ratio(london_house_prices,"Greenwich"))
```

```python
# Add 2018_to_1998_pr column to dataframe
london_house_prices['2018_to_1998_pr'] = london_house_prices.apply(lambda row: 
                                                                    create_price_ratio(london_house_prices,row["london_borough"]),axis=1)
london_house_prices.head()
```

```python
# Create a dataframe boroughs_by_pr that contains the unique list of boroughs
# and their 2018:1998 average price ratio.
# Note: City of London is included in the data set even though
#       it's not technically a borough.
# Sort dataframe by from highest to lowest price ratio.
boroughs_by_pr = london_house_prices[['london_borough','2018_to_1998_pr']].drop_duplicates('london_borough')
boroughs_by_pr = boroughs_by_pr.sort_values('2018_to_1998_pr',ascending=False)
print(boroughs_by_pr)
```
